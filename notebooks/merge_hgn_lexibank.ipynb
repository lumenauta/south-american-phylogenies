# Task
Merge the datasets "lexibank_final.tsv" and "hgn_final.tsv", then sort the merged dataset alphabetically by the 'language' column and then by the 'gloss' column.

## Load the datasets

### Subtask:
Load the `lexibank_final.tsv` and `hgn_final.tsv` datasets into pandas DataFrames.


**Reasoning**:
The subtask is to load two TSV files into pandas DataFrames. This can be done by importing pandas and then using `pd.read_csv` for each file with the tab delimiter.



```python
import pandas as pd

df_lexibank = pd.read_csv('/content/lexibank_final.tsv', delimiter='\t')
df_hgn = pd.read_csv('/content/hgn_final.tsv', delimiter='\t')
```

## Merge the datasets

### Subtask:
Merge the two DataFrames based on common columns.


**Reasoning**:
Merge the two dataframes based on common columns using an outer merge to keep all rows.



```python
common_cols = list(set(df_lexibank.columns) & set(df_hgn.columns))
df_merged = pd.merge(df_lexibank, df_hgn, on=common_cols, how='outer')
display(df_merged.head())
```

## Sort the merged dataset

### Subtask:
Sort the merged DataFrame alphabetically by the 'language' column and then by the 'gloss' column.


**Reasoning**:
Sort the merged DataFrame by the 'LANGUAGE' column and then by the 'GLOSS' column and then display the head of the sorted DataFrame to verify the sorting.



```python
df_merged.sort_values(by=['LANGUAGE', 'GLOSS'], inplace=True)
display(df_merged.head())
```

## Display the result

### Subtask:
Display the first few rows of the merged and sorted DataFrame.


**Reasoning**:
Display the first few rows of the merged and sorted DataFrame as requested by the subtask.



```python
display(df_merged.head())
```

## Remove duplicate rows

### Subtask:
Remove duplicate rows from the merged and sorted DataFrame.

**Reasoning**:
Use the `drop_duplicates()` method to remove duplicate rows from the merged DataFrame and then display the head of the resulting DataFrame to verify the removal of duplicates.

```python
df_merged.drop_duplicates(subset=['CONCEPTICON_GLOSS', 'GLOTTOCODE', 'TOKENS'], inplace=True)
display(df_merged.head())
```
