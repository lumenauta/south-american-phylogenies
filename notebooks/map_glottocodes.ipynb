```python
import pandas as pd

# === 1. Cargar tu lista ===
input_file = "lenguas_old.csv"  # tu archivo
df = pd.read_csv(input_file)

# === 2. Descargar o cargar el archivo de Glottolog ===
# Puedes obtenerlo desde: https://glottolog.org/meta/downloads/languages-and-dialects-geo.csv
glottolog_file = "languages_and_dialects_geo.csv"
glottolog = pd.read_csv(glottolog_file)

# === 3. Hacer el mapeo ===
# Normalizamos nombres para comparar mejor
def normalize(s):
    return str(s).strip().lower().replace("’", "'")

glottolog["name_norm"] = glottolog["name"].apply(normalize)
df["Language_norm"] = df["Language"].apply(normalize)

# Unimos los dataframes
merged = df.merge(glottolog, how="left", left_on="Language_norm", right_on="name_norm")

# === 4. Guardar resultado ===
output = merged[["Language", "Family", "Number of Glosses", "glottocode"]]
output.to_csv("languages_with_glottocodes.csv", index=False)

print("✅ Mapeo completado. Archivo guardado como languages_with_glottocodes.csv")
```
