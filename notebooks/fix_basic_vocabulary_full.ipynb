```python
import pandas as pd

df = pd.read_csv('/content/basic_vocabulary_full.csv')
df = df.reset_index()
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
columns_to_drop = ['Source', 'Language', 'ID', 'GLOSS', 'Orthographic Form', 'Phonemicized Form', 'Gloss as in Source', 'Etymology Code', 'Proto-Language', 'Loan Source', 'Wanderwort Status', 'Etyma Set', 'Etymology Notes', 'General Notes']
df = df.drop(columns=columns_to_drop)
display(df.head())
```

```python
empty_cells_per_column = df.isnull().sum()
print("Number of empty cells per column:")
print(empty_cells_per_column)
```

```python
df = df.dropna(subset=['Proto-Form'])
display(df.head())
```

```python
df_split = df['Gloss Link'].str.split('/', expand=True).stack().reset_index(level=1, drop=True).to_frame('Gloss Link')
df_split = df_split.merge(df.drop('Gloss Link', axis=1), left_index=True, right_index=True, how='left')
# Reorder columns to match the original DataFrame
original_cols = df.columns.tolist()
df_split = df_split[original_cols]
display(df_split.head())
```

```python
df_split['Proto-Form'] = df_split['Proto-Form'].str.replace(r" \(.+\)(?=[,;])", "", regex=True)
display(df_split.head())
```

```python
df_split['Proto-Form'] = df_split['Proto-Form'].str.replace(r" \(.+\)", "", regex=True)
display(df_split.head())
```

```python
df_split['Proto-Form'] = df_split['Proto-Form'].str.replace(r" \[.+\](?=[,;])", "", regex=True)
display(df_split.head())
```

```python
df_split['Proto-Form'] = df_split['Proto-Form'].str.replace(r" \[.+\]", "", regex=True)
display(df_split.head())
```

```python
df_split.to_csv('cleaned_basic_vocabulary.csv', index=False)
```
