# Task
Match rows in "merged_vocabulary_with_glottocodes.csv" to "glosas_unicas_hgn-mapped.tsv" based on the 'GLOSS' column and add the corresponding 'CONCEPTICON_GLOSS' and 'CONCEPTICON_ID' from the second file.

## Load data

### Subtask:
Load both `merged_vocabulary_with_glottocodes.csv` and `glosas_unicas_hgn-mapped.tsv` into pandas DataFrames.


**Reasoning**:
Load the data into pandas DataFrames as requested in the instructions.



```python
import pandas as pd

df_vocabulary = pd.read_csv('/content/merged_vocabulary_with_glottocodes.csv')
df_mapping = pd.read_csv('/content/glosas_unicas_hgn-mapped.tsv', sep='\t')
```

## Prepare mapping data

### Subtask:
Select the relevant columns (`GLOSS`, `CONCEPTICON_GLOSS`, and `CONCEPTICON_ID`) from the `glosas_unicas_hgn-mapped.tsv` DataFrame and rename the `GLOSS` column to match the column in the first DataFrame for merging.


**Reasoning**:
Select the relevant columns from `df_mapping` and rename the 'GLOSS' column to ensure consistency for merging.



```python
df_mapping_selected = df_mapping[['GLOSS', 'CONCEPTICON_GLOSS', 'CONCEPTICON_ID']]
df_mapping_selected.rename(columns={'GLOSS': 'GLOSS'}, inplace=True)
```

## Merge dataframes

### Subtask:
Merge the two DataFrames based on the common GLOSS column.


**Reasoning**:
Merge the two dataframes based on the common 'GLOSS' column using a left merge.



```python
df_merged = pd.merge(df_vocabulary, df_mapping_selected, on='GLOSS', how='left')
display(df_merged.head())
```

## Handle unmatched rows

### Subtask:
Identify and handle any rows from `merged_vocabulary_with_glottocodes.csv` that did not find a match in `glosas_unicas_hgn-mapped.tsv`.


**Reasoning**:
Identify rows in df_merged where CONCEPTICON_ID is missing and display the count of such rows.



```python
unmatched_rows = df_merged[df_merged['CONCEPTICON_ID'].isna()]
print(f"Number of unmatched rows: {len(unmatched_rows)}")
display(unmatched_rows.head())
```

## Save results

### Subtask:
Save the merged DataFrame to a new CSV file.


**Reasoning**:
Save the merged DataFrame to a CSV file without the index.



```python
df_merged = df_merged[['GLOSS', 'CONCEPTICON_GLOSS', 'CONCEPTICON_ID', 'LANGUAGE', 'glottocode', 'FAMILY', 'FORM', 'TOKENS']]
df_merged.to_csv('merged_vocabulary_with_concepticons.csv', index=False)
```

### Data Analysis Key Findings

*   The dataset `glosas_unicas_hgn-mapped.tsv` contained mappings between glosses and Concepticon information (`CONCEPTICON_GLOSS` and `CONCEPTICON_ID`).
*   Merging the two datasets on the 'GLOSS' column resulted in a combined dataset where Concepticon information was added to the vocabulary entries where a match was found.
*   There were 1062 rows in the original vocabulary data that did not have a corresponding 'GLOSS' entry in the mapping file, resulting in missing (`NaN`) values for `CONCEPTICON_GLOSS` and `CONCEPTICON_ID` in the merged dataset.
