# Task
Map the `CONCEPTICON_ID` and `CONCEPTICON_GLOSS` columns from "p-arawakan-clean-mapped.tsv" onto "p-arawakan.tsv" by joining on the `GLOSS` column. Comment on the cases that could not be mapped.

## Load data

### Subtask:
Load both `p-arawakan.tsv` and `p-arawakan-clean-mapped.tsv` into pandas DataFrames.


**Reasoning**:
Import pandas and load the two specified TSV files into pandas DataFrames.



```python
import pandas as pd

name = 'p-zaparoan'
df_arawakan = pd.read_csv(f'{name}.tsv', sep='\t')
df_mapped = pd.read_csv(f'{name}-clean-mapped.tsv', sep='\t')
```

**Reasoning**:
Display the first few rows of both dataframes to verify they were loaded correctly.



```python
display(df_arawakan.head())
display(df_mapped.head())
```

## Merge dataframes

### Subtask:
Merge the two DataFrames on the `GLOSS` column, keeping all rows from `p-arawakan.tsv`.


**Reasoning**:
Merge the two dataframes using a left merge on the 'GLOSS' column and display the first few rows of the resulting dataframe.



```python
df_merged = pd.merge(df_arawakan, df_mapped[['GLOSS', 'CONCEPTICON_ID', 'CONCEPTICON_GLOSS']], on='GLOSS', how='left')
display(df_merged.head())
```

## Handle unmapped cases

### Subtask:
Identify the rows in the merged DataFrame where `CONCEPTICON_ID` and `CONCEPTICON_GLOSS` are missing (indicating they couldn't be mapped).


**Reasoning**:
Filter the merged DataFrame to find rows where the concept icon information is missing.



```python
df_unmapped = df_merged[df_merged['CONCEPTICON_ID'].isnull()]
display(df_unmapped.head())
```

## Comment on unmapped cases

### Subtask:
Provide a summary or display the rows that could not be mapped.


**Reasoning**:
Print a statement and display the unmapped rows as requested by the subtask.



```python
print("The following rows could not be mapped:")
display(df_unmapped)
```

```python
# Filter out the rows that could not be mapped
df_mapped_only = df_merged.dropna(subset=['CONCEPTICON_ID', 'CONCEPTICON_GLOSS'])

print("\nFirst few rows of the dataframe with only mapped cases:")
display(df_mapped_only.head())
```

## Save results

### Subtask:
Optionally save the merged DataFrame to a new file.


**Reasoning**:
Save the merged DataFrame to a TSV file without the index.



```python
# Define the desired column order
new_column_order = ['GLOSS', 'CONCEPTICON_GLOSS', 'CONCEPTICON_ID', 'LANGUAGE', 'GLOTTOCODE', 'FAMILY', 'FORM']

# Reindex the DataFrame with the new column order
df_merged_reordered = df_merged.reindex(columns=new_column_order)

# Save the reordered DataFrame to a TSV file without the index.
df_merged_reordered.to_csv(f'{name}-concepticon.tsv', sep='\t', index=False)
```

### Data Analysis Key Findings

*   Rows where the `GLOSS` value in `p-arawakan.tsv` did not have a corresponding match in `p-arawakan-clean-mapped.tsv` resulted in `NaN` values in the `CONCEPTICON_ID` and `CONCEPTICON_GLOSS` columns in the merged dataframe.
