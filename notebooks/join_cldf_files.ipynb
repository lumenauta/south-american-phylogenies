```python
import pandas as pd

file_path = 'forms.csv'
df = pd.read_csv(file_path, sep=',', engine='python')
display(df.head())

file_path = 'languages.csv'
df = pd.read_csv(file_path, sep=',', engine='python')
display(df.head())

file_path = 'parameters.csv'
df = pd.read_csv(file_path, sep=',', engine='python')
display(df.head())
```

```python
# Assuming the dataframes are already loaded as df_forms, df_languages, and df_parameters
# If not, load them first
df_forms = pd.read_csv('forms.csv', sep=',', engine='python')
df_languages = pd.read_csv('languages.csv', sep=',', engine='python')
df_parameters = pd.read_csv('parameters.csv', sep=',', engine='python')
#df_cognates = pd.read_csv('cognates.csv', sep=',', engine='python')

# Join df_forms and df_languages on forms.Language_ID = languages.ID
merged_df = pd.merge(df_forms, df_languages, left_on='Language_ID', right_on='ID', how='inner', suffixes=('_forms', '_languages'))

# Join the result with df_parameters on forms.Parameter_ID = parameters.ID
final_df = pd.merge(merged_df, df_parameters, left_on='Parameter_ID', right_on='ID', how='inner', suffixes=('', '_parameters'))

# Display the first few rows of the final joined dataframe
pd.set_option('display.max_columns', None)
display(final_df.head())
display(final_df.tail())
```

```python
print(final_df.columns.tolist())
```

```python
columns_to_keep = [
    'Form',
    'Segments',
    'Name',
    'Glottocode',
    'Glottolog_Name',
    'Family',
    'Name_parameters',
    'Concepticon_ID',
    'Concepticon_Gloss'
]

# Create a new DataFrame with only the specified columns
df_subset = final_df[columns_to_keep].copy()

# Display the first few rows of the new DataFrame
display(df_subset.head())
```

```python
# Define the new order and names for the columns
new_column_order_and_names = {
    'Name_parameters': 'GLOSS',
    'Concepticon_Gloss': 'CONCEPTICON_GLOSS',
    'Concepticon_ID': 'CONCEPTICON_ID',
    'Name': 'LANGUAGE',
    'Glottocode': 'GLOTTOCODE',
    'Family': 'FAMILY',
    'Form': 'FORM',
    'Segments': 'TOKENS'
}

# Select, reorder, and rename the columns
df_reordered_renamed = df_subset[list(new_column_order_and_names.keys())].rename(columns=new_column_order_and_names)

# Display the first few rows of the new DataFrame
display(df_reordered_renamed.head())
```

```python
# Save the DataFrame as a TSV file
df_reordered_renamed.to_csv('walkerarawakan.tsv', sep='\t', index=False)
```
