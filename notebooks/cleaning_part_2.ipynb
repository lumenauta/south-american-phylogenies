```python
import pandas as pd

# Load the dataframe.
df = pd.read_csv('/content/cleaned_basic_vocabulary_int.csv')

# Drop rows containing '#NAME?'.
df = df[~df.isin(['#NAME?']).any(axis=1)]

# Display the head of the dataframe.
display(df.head())
```

```python
# Save the modified dataframe to a new CSV file.
df.to_csv('/content/cleaned_basic_vocabulary_int_2.csv', index=False)

print("DataFrame saved successfully to '/content/cleaned_df.csv'")
```
