```python
import pandas as pd

try:
    df = pd.read_csv('cleaned_basic_vocabulary_int.csv', sep=',', on_bad_lines='skip')
    display(df.head())
except Exception as e:
    print(f"An error occurred while reading the CSV: {e}")
```

```python
new_rows = []
for index, row in df.iterrows():
    proto_forms = row['Form']
    # Check if proto_forms is a string before splitting
    if isinstance(proto_forms, str):
        # Split by semicolon, tilde, and slash first, then by comma
        split_forms = []
        # Split by semicolon
        semicolon_split = proto_forms.split(';')
        for item in semicolon_split:
            # Split by tilde
            tilde_split = item.split('~')
            for sub_item in tilde_split:
                # Split by slash
                slash_split = sub_item.split('/')
                for sub_sub_item in slash_split:
                    split_forms.extend(sub_sub_item.split(','))


        for form in split_forms:
            # Remove leading/trailing whitespace and quotes
            cleaned_form = form.strip().strip('"')
            new_row = row.to_dict()
            new_row['Form'] = cleaned_form
            new_rows.append(new_row)
    else:
        # If not a string, just append the original row
        new_rows.append(row.to_dict())

processed_df = pd.DataFrame(new_rows)
display(processed_df.head())
```

```python
file_path = 'cleaned_basic_vocabulary_int.csv'

try:
    # Get the total number of lines in the original file
    with open(file_path, 'r', encoding='utf-8') as f:
        total_lines = sum(1 for line in f)

    # Get the number of rows in the DataFrame loaded with skipped lines
    # The DataFrame 'df' was created using on_bad_lines='skip'
    num_rows_in_df = len(df)

    skipped_lines_count = total_lines - num_rows_in_df

    print(f"Total lines in the original file: {total_lines}")
    print(f"Number of rows in the DataFrame (after skipping bad lines): {num_rows_in_df}")
    print(f"Number of lines skipped: {skipped_lines_count}")

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
processed_df.to_csv('processed_vocabulary.csv', index=False)
```
