```python
import pandas as pd

# Load the data from the CSV file
try:
    df = pd.read_csv("processed_vocabulary_cleaned.csv", on_bad_lines='skip')
except FileNotFoundError:
    print("Error: processed_vocabulary_cleaned.csv not found. Please make sure the file is in the correct directory.")
    # exit() # Removed exit() to allow the rest of the code to execute if the file is found
```

```python
# Check if the DataFrame is loaded and has at least 4 columns
if 'df' in locals() and len(df.columns) >= 4:
    # Filter rows where the 4th column (index 3) contains '-'
    rows_with_char = df[df.iloc[:, 3].astype(str).str.contains(' ')]

    # Display the filtered rows
    if not rows_with_char.empty:
        print("Rows where the 4th column contains ' ':")
        display(rows_with_char)
    else:
        print("No rows found where the 4th column contains ' '.")
elif 'df' not in locals():
    print("DataFrame could not be loaded. Please run the cell to load the data first.")
else:
    print("The DataFrame does not have at least 4 columns.")
```

```python
from ipapy import is_valid_ipa

# Define a function to check for non-IPA characters in a string, excluding rows with only newline as non-IPA
def contains_non_ipa_proper(text):
    if not isinstance(text, str):
        return False  # Return False for non-string values

    # Use ipapy's function to check for valid IPA characters
    # This function checks the entire string. We want to know if there are *any* non-IPA chars other than newline
    # A simple approach is to check if the string without newlines is not considered valid IPA
    text_without_newlines = text.replace('\n', '')
    if not text_without_newlines: # if only newlines remain, consider it valid for this purpose
        return False

    return not is_valid_ipa(text_without_newlines)


# Check if the DataFrame is loaded and has at least 4 columns
if 'df' in locals() and len(df.columns) >= 4:
    # Apply the function to the 4th column to create a boolean mask
    non_ipa_mask_proper = df.iloc[:, 3].astype(str).apply(contains_non_ipa_proper)

    # Filter rows where the 4th column contains non-IPA characters (excluding those with only newline)
    rows_with_non_ipa_proper = df[non_ipa_mask_proper]

    # Display the filtered rows
    if not rows_with_non_ipa_proper.empty:
        print("Rows where the 4th column contains non-IPA characters (properly checked, excluding only newline):")
        display(rows_with_non_ipa_proper)
    else:
        print("No rows found where the 4th column contains non-IPA characters (properly checked, excluding only newline).")
elif 'df' not in locals():
    print("DataFrame could not be loaded. Please run the cell to load the data first.")
else:
    print("The DataFrame does not have at least 4 columns.")
```

```python
# Check if the DataFrame rows_with_non_ipa_proper is loaded
if 'rows_with_non_ipa_proper' in locals():
    # Get all unique characters from the 4th column of the filtered rows
    all_chars_in_filtered_rows = set(''.join(rows_with_non_ipa_proper.iloc[:, 3].astype(str).tolist()))

    # Sort the unique characters
    sorted_unique_chars = sorted(list(all_chars_in_filtered_rows))

    if sorted_unique_chars:
        print("Unique characters found in the 4th column of rows identified as containing non-IPA characters:")
        print(sorted_unique_chars)
    else:
        print("No characters found in the 4th column of the filtered rows.")
elif 'df' not in locals():
     print("DataFrame could not be loaded. Please run the cell to load the data first.")
else:
    print("The filtered DataFrame 'rows_with_non_ipa_proper' does not exist. Please run the cell that creates it first.")
```
