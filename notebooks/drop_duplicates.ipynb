```python
import pandas as pd

df_merged = pd.read_csv('/content/protoforms_final.tsv', delimiter='\t')
```

```python
df_merged.sort_values(by=['LANGUAGE', 'GLOSS'], inplace=True)
display(df_merged.head())
```

```python
df_original = df_merged.copy()
df_merged.drop_duplicates(subset=['CONCEPTICON_GLOSS', 'GLOTTOCODE', 'TOKENS'], inplace=True)

df_dropped = df_original[~df_original.index.isin(df_merged.index)]

display(df_merged.head())
```

```python
df_merged.to_csv('merged_sorted_dataset.tsv', index=False, sep='\t')
df_dropped.to_csv('protoforms_dropped.tsv', index=False, sep='\t')
```
