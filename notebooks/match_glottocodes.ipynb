# Task
Match each language in the second column of "processed_vocabulary_segmented_cleaned.csv" with its corresponding glottocode from "languages_with_glottocodes.csv" and save the result to a new CSV file.

## Load data

### Subtask:
Load the `processed_vocabulary_segmented_cleaned.csv` and `languages_with_glottocodes.csv` files into pandas DataFrames.


**Reasoning**:
Import pandas and load the two CSV files into dataframes as instructed.



```python
import pandas as pd

df_vocabulary = pd.read_csv('/content/processed_vocabulary_segmented_cleaned.csv')
df_glottocodes = pd.read_csv('/content/languages_with_glottocodes.csv')

display(df_vocabulary.head())
display(df_glottocodes.head())
```

## Clean data

### Subtask:
Clean the language names in both DataFrames to ensure consistent formatting for accurate matching. This may involve removing leading/trailing spaces, standardizing capitalization, or handling variations in language names.


**Reasoning**:
Convert language names to lowercase and remove leading/trailing whitespace in both dataframes.



```python
df_vocabulary['LANGUAGE'] = df_vocabulary['LANGUAGE'].str.lower().str.strip()
df_glottocodes['Language'] = df_glottocodes['Language'].str.lower().str.strip()

display(df_vocabulary.head())
display(df_glottocodes.head())
```

## Match languages

### Subtask:
Merge the two DataFrames based on the cleaned language names to associate each entry in the vocabulary data with its corresponding glottocode.


**Reasoning**:
Perform a left merge of df_vocabulary with df_glottocodes on the cleaned language name columns and display the head of the merged DataFrame.



```python
df_merged = pd.merge(df_vocabulary, df_glottocodes[['Language', 'glottocode']], left_on='LANGUAGE', right_on='Language', how='left')
df_merged = df_merged.drop('Language', axis=1)
display(df_merged.head())
```

## Handle unmatched languages

### Subtask:
Identify any languages in the vocabulary data that could not be matched with a glottocode and decide on a strategy for handling them (e.g., marking them as unmatched or attempting further cleaning/matching).


**Reasoning**:
Identify and display unique languages with missing glottocodes and count the total number of entries with missing glottocodes.



```python
unmatched_languages = df_merged[df_merged['glottocode'].isnull()]['LANGUAGE'].unique()
print("Languages in vocabulary data that could not be matched with a glottocode:")
print(unmatched_languages)
print("\nNumber of entries with missing glottocodes:", df_merged['glottocode'].isnull().sum())
```

**Reasoning**:
Briefly consider potential reasons for the unmatched languages based on the list of unmatched languages.



```python
print("\nPotential reasons for unmatched languages:")
print("- Variations in language names or spelling (e.g., 'candoshishapra' vs. a potentially different spelling in the glottocodes data).")
print("- Compound names or dialects that might be listed differently (e.g., 'moseténchimané' or 'ese ejja/huarayo').")
print("- Languages present in the vocabulary data but not included in the glottocodes data.")
print("- Errors in data entry in either dataset.")
```

## Save results

### Subtask:
Save the merged DataFrame with the added glottocode column to a new CSV file.


**Reasoning**:
Save the merged DataFrame to a CSV file without the index.



```python
df_merged.to_csv('merged_vocabulary_with_glottocodes.csv', index=False)
```
