```python
import pandas as pd

name = 'p-zaparoan'
df = pd.read_csv(f'{name}.csv')
#df = df.drop('index', axis=1)
df = df.dropna(subset=['FORM'])
df['Language'] = 'Proto-Zaparoan'
df['Glottocode'] = 'zapa1251'
df['Family'] = 'Zaparoan'
display(df.head())
```

```python
s_gloss = df['Gloss'].str.replace('"', '').str.split(',').explode()
s_gloss.name = 'Gloss'
df = df.drop('Gloss', axis=1).join(s_gloss)

# Remove text within parentheses from the 'Gloss' column
df['Gloss'] = df['Gloss'].str.replace(r'\(.*?\)', '', regex=True).str.strip()

display(df.head())
```

```python
# Display columns of df right before the line causing the error
print("Columns of df before processing 'Phonemicized Form':")
print(df.columns)

s_form = df['FORM'].str.split(r'[~,]').explode()
s_form.name = 'FORM'
df = df.drop('FORM', axis=1).join(s_form)

# Remove unwanted characters from the FORM column
df['FORM'] = df['FORM'].str.replace(r'[*\-\(\)]', '', regex=True)

display(df.head())
```

```python
#df = df[['Gloss', 'Language', 'Glottocode', 'Family', 'Phonemicized Form']]
df = df[['Gloss', 'Language', 'Glottocode', 'Family', 'FORM']]

df.to_csv(f'{name}.tsv', sep='\t', index=False, quoting=3)
```
