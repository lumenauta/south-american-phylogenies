```python
import csv

with open("basic_vocabulary_full.csv", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for i, row in enumerate(reader, start=1):
        gloss = row["Gloss"].strip()
        if gloss == "" or gloss in ["-", "–"]:
            print(f"Fila {i} tiene glos vacía o inválida: '{row['GLOSS']}'")
```
