# Task
Create Python code to load the CSV file "processed_vocabulary.csv" into a dataframe, remove all parentheses and whitespace characters from the 4th column, and display the updated dataframe.

## Load the data

### Subtask:
Load the `processed_vocabulary.csv` file into a pandas DataFrame.


**Reasoning**:
Load the CSV file into a pandas DataFrame.



```python
import pandas as pd
df = pd.read_csv('processed_vocabulary.csv')
```

**Reasoning**:
Display the first few rows of the loaded DataFrame to verify the data was loaded correctly.



```python
df.head()
```

## Process the data

### Subtask:
Strip parentheses and whitespace characters from the 4th column.


**Reasoning**:
Strip parentheses and whitespace characters from the 4th column.



```python
df.iloc[:, 3] = df.iloc[:, 3].str.replace('[() ]', '', regex=True).str.strip()
display(df)
```

```python
df.iloc[:, 3] = df.iloc[:, 3].str.replace('[1-9]', '', regex=True).str.strip()
display(df)
```

## Display the result

### Subtask:
Display the modified DataFrame to show the changes.


**Reasoning**:
Display the DataFrame to show the changes made in the previous step.



```python
display(df)
```
