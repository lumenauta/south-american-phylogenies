Merge the `languages_and_dialects_geo.csv` geographical data into `merged_final.tsv` by matching on 'glottocode'/'DOCULECT', and then confirm that 'latitude' and 'longitude' columns have been successfully added to the merged data.

## Load Data

### Subtask:
Load 'merged_final.tsv' and 'languages_and_dialects_geo.csv' into pandas DataFrames.


**Reasoning**:
Import the pandas library and then load the two specified files, 'merged_final.tsv' and 'languages_and_dialects_geo.csv', into pandas DataFrames as instructed.



```python
import pandas as pd

df_merged_final = pd.read_csv('/content/merged_final.tsv', delimiter='\t')
df_geo = pd.read_csv('/content/languages_and_dialects_geo.csv')

print("df_merged_final loaded successfully. First 5 rows:")
print(df_merged_final.head())
print("\ndf_geo loaded successfully. First 5 rows:")
print(df_geo.head())
```

## Prepare Geo Data

### Subtask:
Rename the 'glottocode' column in the 'languages_and_dialects_geo.csv' DataFrame to 'DOCULECT' to match the column in 'merged_final.tsv' for joining.


**Reasoning**:
Rename the 'glottocode' column to 'DOCULECT' in `df_geo` to prepare for merging and verify the change by displaying the first 5 rows and column names.



```python
df_geo = df_geo.rename(columns={'glottocode': 'DOCULECT'})

print("df_geo after renaming 'glottocode' to 'DOCULECT':")
print(df_geo.head())
print("\nUpdated columns of df_geo:")
print(df_geo.columns)
```

**Reasoning**:
Now that the 'glottocode' column in `df_geo` has been renamed to 'DOCULECT', the next step is to merge `df_geo` into `df_merged_final` based on the common 'DOCULECT' column. A left merge will ensure all data from `df_merged_final` is retained, and matching geographical data is added.



```python
df_merged_final = pd.merge(df_merged_final, df_geo[['DOCULECT', 'latitude', 'longitude']], on='DOCULECT', how='left')

print("df_merged_final after merging geographical data. First 5 rows:")
print(df_merged_final.head())
print("\nColumns of df_merged_final after merge:")
print(df_merged_final.columns)
```
