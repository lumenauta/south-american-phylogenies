```python
import csv
import re

def clean_gloss(gloss: str) -> str:
    gloss = gloss.strip().lower()
    gloss = gloss.replace('‘', '').replace('’', '').replace('“', '').replace('”', '')
    gloss = gloss.strip('\'"')  # comillas normales
    gloss = gloss.lower()
    gloss = re.sub(r"\(.*?\)", "", gloss)
    gloss = gloss.replace("?", "")
    gloss = re.sub(r"\s+", " ", gloss)
    return gloss.strip()

def prepare_wordlist(input_csv, output_csv):
    with open(input_csv, encoding="utf-8") as f_in, open(output_csv, "w", encoding="utf-8", newline="") as f_out:
        reader = csv.reader(f_in)
        headers = next(reader)  # leer la primera fila
        # buscar columna que contenga 'gloss' (ignora mayúsculas/minúsculas)
        gloss_idx = None
        for i, h in enumerate(headers):
            if 'gloss' in h.lower():
                gloss_idx = i
                break
        if gloss_idx is None:
            raise ValueError(f"No se encontró columna de glosas en {input_csv}")

        writer = csv.writer(f_out)
        writer.writerow(["GLOSS"])

        for row in reader:
            raw_gloss = row[gloss_idx] if gloss_idx < len(row) else ""
            if not raw_gloss:
                continue
            # dividir glosas compuestas
            for part in re.split(r"[;,/]", raw_gloss):
                gloss = clean_gloss(part)
                # ignorar sufijos/morfemas
                if gloss and not gloss.startswith("*") and not gloss.startswith("-"):
                    writer.writerow([gloss])
```

```python
wordlists = ["p-arawakan", 'p-aymaran', 'p-chibchan', 'p-chapacuran', 'p-chicham', 'p-chocoan', 'p-guahiban', 'p-guaicuru', 'p-macro-je', 'p-matacoan', 'p-nadahupan', 'p-panoan', 'p-quechuan', 'p-zaparoan']

if __name__ == "__main__":
  for f in wordlists:
    prepare_wordlist(f'{f}.csv', f'{f}-clean.csv')
    files.download(f"{f}-clean.csv")
    print(f"✅ Archivo preparado: {f}-clean.csv")
```
